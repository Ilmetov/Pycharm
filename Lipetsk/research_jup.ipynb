```python
import pandas as pd
import numpy as np

cht = pd.read_csv('data/chg_tariffs.csv', sep=';')
cht['DATE'] = pd.to_datetime(cht['DATE']) #преобразуем в тип данных дата и время
cht = cht.sort_values(by=['uid', 'DATE'], ascending=[True, True]) #делаем сортировку, чтобы визуально убедиться
cht['NEXT_DATE'] = cht.groupby(['uid'])['DATE'].shift(-1) #подтягиваем в разрезе каждого абонента данные о следующей дате перехода, чтобы посчитать время пребывания абонента в одном статусе  
cht = cht.assign(TIME_DIFF=cht.NEXT_DATE - cht.DATE) #посчитаем сколько времени абонент провел на одном тарифе
cht = cht.assign(MONEY_DIFF=cht.NEW_PRICE - cht.OLD_PRICE) #посчитаем увеличил ли он тариф или уменьшил
def new(row):
    if row['OLD_PRICE'] == 0:
        return 'Y'
    else:
        return 'N'

def out(row):
    if row['NEW_PRICE'] == 0:
        return 'Y'
    else:
        return 'N'
cht['NEW_FLAG'] = cht.apply(new, axis=1)
cht['OUT_FLAG'] = cht.apply(out, axis=1)
cht #написать про косяки данных (клиенты, которые появляются посреди своего жизненного цикла, или которые открывают счета несколько раз), перечитать ещё раз предложения Александра, дать комментарии, проверить выбытие абонентов по таблице с абонентами,  
```
